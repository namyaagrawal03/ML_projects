```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
```

```python
df_features=df[['Movie_Genre', 'Movie_Keywords' ,'Movie_Overview', 'Movie_Cast' ,'Movie_Director', 'Movie_Tagline']].fillna(' ')
```

```python
df_features.shape
```

```python
df_features
```

```python
X = df_features['Movie_Cast'] + ' ' + df_features['Movie_Director'] + ' ' + df_features['Movie_Genre'] + ' ' + df_features['Movie_Keywords']+ ' ' + df_features['Movie_Overview']+ ' ' +df_features['Movie_Tagline']
```

```python
X.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X= tfidf.fit_transform(X)
```

```python
X.shape
```

```python
print(X)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(X)
similarity_score
```

```python
similarity_score.shape
```

```python
movie_input= input('Enter the movie name: ')
all_movies= df['Movie_Title'].tolist()
#for closest spelling
import difflib
Movie_rec = difflib.get_close_matches(movie_input, all_movies)
close_match=Movie_rec[0]
index_of_close_match= df[df.Movie_Title==close_match]['Movie_ID'].values[0]
#getting a list of similar movies
rec_score= list(enumerate(similarity_score[index_of_close_match]))
#get all movies based on recommendation score wrt input movie

#sorting movies based on similarity score

sorted_sim_movies=sorted(rec_score, key= lambda x:x[1], reverse= True)
print("Top 10 movies for you:" )
i=1

for movie in sorted_sim_movies:
  index= movie[0]
  title_from_index= df[df.index==index]['Movie_Title'].values[0]
  if (i<11):
    print(i, ".", title_from_index)
    i+=1
```
